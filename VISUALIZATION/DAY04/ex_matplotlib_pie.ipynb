### [박스 그래프]
- 컬럼(변수)의 이상치 데이터 및 분포 확인

```python
# [1] 모듈로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager as fm
from matplotlib import rc
```

```python
# [2] 데이터 준비
F_DATA=r'C:\KDT\EX_PANDAS\DATAS\auto_mpg.csv'

# 적용할 폰트 파일
FONT_FILE=r'C:\Windows\Fonts\NanumGothic.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=FONT_FILE).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# [3] CSV => DataFrame으로 저장
dataDF=pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율 시각화
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
originSR=dataDF['origin'].value_counts()
x=originSR.index.to_list()
h=originSR.values.tolist()
x, h
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar 그래프
originBar=plt.bar(x,h,color=['red','green','blue'], edgecolor=['black'])

# 막대에 라벨을 설정
plt.bar_label(originBar,h, label_type='center')

# x축 눈금 설정
plt.xticks([1,2,3],['미국','유럽','일본'])
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['미국','유럽','일본'], autopct='%.1f%%')
plt.show()
```

```python
plt.pie(h,labels=['미국','유럽','일본'], autopct='%.1f%%',
        counterclock=False, startangle=90,explode=[0,0,0.1234])
plt.show()
```

```python
plt.pie(h,labels=['미국','유럽','일본'], autopct='%.1f%%',
        counterclock=False, startangle=90, wedgeprops={'linewidth':1, 'width':0.5, 'edgecolor':'w'})
plt.show()
```
