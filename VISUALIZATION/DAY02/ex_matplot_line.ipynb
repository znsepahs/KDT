### [Matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈로딩
# - 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(y축 데이터)
plt.plot([1,2,3,4])
plt.xlabel("DATA")
plt.ylabel("RESULT")
plt.show()
```

```python
# [2] 그래프 그리기 - Line (선) 그래프
# plot(x축 데이터, y축 데이터)
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
# [2] 그래프 그리기 - Line (선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포맷스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'r--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line (선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포맷스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'y<--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line (선) 그래프
# plot(x축 데이터, y축 데이터, line2D속성으로 설정)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],color='green',marker='o',linestyle='dashed',
         markersize=10,linewidth=4,markeredgecolor='yellow',markeredgewidth='5')
plt.show()
```

```python
# [2] 그래프 그리기 - Line (선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포맷스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
# plt.plot([10,20,30],[15,25,35],'y<-')
# plt.plot([10,20,30],[18,29,25],'r<--')
# plt.plot([10,20,30],[25,25,30],'k^-') / 이렇게도 가능하고 아래 처럼 한줄로 코드 작성 가능
plt.plot([10,20,30],[15,25,35],'y<-',[10,20,30],[18,29,25],'r<--',[10,20,30],[25,25,30],'k^-')
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'bo--',label="V")
plt.plot([10,20,30],[5,30,28],'yo--',label="Z")
plt.title('[MPLIB LINE2D]')
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함
#                            loc속성으로 위치 지정 가능
plt.plot([10,20,30],[15,25,20],'bo--',label="V")
plt.plot([10,20,30],[5,30,28],'yo--',label="Z")
plt.title('[MPLIB LINE2D]')
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.legend(loc='lower right')
plt.show()
```
