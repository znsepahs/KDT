```python
# [1-1] 모듈로딩 
import pandas as pd                     # 데이터 분석용 모듈
import matplotlib.pyplot as plt         # 데이터 시각화용 모듈
```

```python
# [1-2] 데이터 준비
F_DATA=r'C:\KDT\EX_PANDAS\DATAS\시도별 전출입 인구수.xlsx'
```

```python
# [1-3] DATA ==> DataFrame 형태 저장
dataDF=pd.read_excel(F_DATA, engine='openpyxl')
```

- [3] 데이터 전처리 
    * 기본확인 파악된 내용들 처리
    * 불필요 데이터 삭제 : 0번행 삭제
    * 결측치 데이터 처리 : '전출지별'컬럼 처리
    * 이상치 데이터 처리 : '-'에대한 처리
    * 데이터 자료형 일치 : 형변환

```python
# [3-1] 불필요 데이터 처리
copyDF=dataDF.drop(index=0)
print(f'확인\n {copyDF.head(3)}')
```

```python
# [3-2] 결측치 데이터 처리 => isna()/isnull()
copyDF.isna().sum()
#print(f'확인\n {copyDF.head(3)}')
```

```python
# - '1984','1985','1986' 컬럼에서 결측치 데이터만 추출 
#print(copyDF.columns)
a=copyDF[['1984','1985','1986']].isna()
a[a.sum(axis=1)>0]
#a.loc[a[1984].isna() | a[1985].isna(), :]
```

```python
# - '전출지별' 
copyDF['전출지별']=copyDF['전출지별'].ffill()

# - 수치데이터 결측치 처리
#pd.set_option('future.no_silent_downcasting', True)
copyDF=copyDF.fillna(0)
```

```python
copyDF.isna().sum()
```

```python
# [3-3] 이상치 데이터 처리 
copyDF.replace('-', 0, inplace=True)
```

```python
copyDF.shape
```

```python
# [3-4] 자료형 변환 : 48개 컬럼을 int로 형변환
d_ = dict(zip(copyDF.columns[2:].to_list(), ['int']*(copyDF.shape[1]-2)))
print(d_)
copyDF=copyDF.astype(d_)
```

```python
print(f'copyDF.dtypes\n {copyDF.dtypes}')
```

- [4] 데이터 가공 
    * 대구에서 전지역 전입 인구수 분석
    * 전출지별 - '대구광역시'
    * 전입지별 - '대구광역시', '전국' 제외

```python
# - 대구광역시 데이터만 추출
mask = (copyDF['전출지별']=='대구광역시')
cityDF=copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF=cityDF.drop('전출지별', axis=1)

# - 전입지별 컬럼 행 인덱스 설정
cityDF.set_index('전입지별', inplace=True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['대구광역시', '전국'], inplace=True)
```

```python
# - 가공데이터 확인
print(cityDF.dtypes, cityDF.columns, cityDF.index, sep='\n\n')
```


- [5] 분석 :  대구에서 전지역 전입 인구수 분석
    * [5-1] 대구 & 타지역 이동 인구수 

```python
# - 데이터 추출
moveTotalDF=cityDF.sum(axis=1)
print(moveTotalDF.index)
```

```python
#- 데이터 시각화 
# (1) 한글 처리
# 한글폰트 설정 => 폰트 매니저 모듈
from matplotlib import font_manager as fm 
from matplotlib import rc 

# 적용할 폰트 파일
FONT_FILE = r'C:\Windows\Fonts\NanumGothic.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=FONT_FILE).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font', family=font_name)
```

```python
fig=plt.figure(figsize=(12,10))
ax1, ax2=fig.subplots(1,2, sharey=True)

ax1.plot(moveTotalDF[:7])
import numpy as np
ax1.set_xticks(np.arange(0,7,1), moveTotalDF.index[:7].to_list(),rotation=45)
#ax1.set_xticklabels(moveTotalDF.index[:7].to_list(), rotation=45)

ax2.bar(moveTotalDF.index[:7].to_list(), moveTotalDF[:7])
ax2.set_xticks(np.arange(0,7,1), moveTotalDF.index[:7].to_list(),rotation=45)

fig.suptitle('[대구광역시 ==> 타광역시 이동인구수]', fontsize=20)
plt.show()
```

```python
fig=plt.figure(figsize=(12,10))
ax1, ax2=fig.subplots(1,2, sharey=True)

ax1.plot(moveTotalDF[7:])
import numpy as np
ax1.set_xticks(np.arange(0,9,1), moveTotalDF.index[7:].to_list(),rotation=45)
#ax1.set_xticklabels(moveTotalDF.index[7:].to_list(), rotation=45)

ax2.bar(moveTotalDF.index[7:].to_list(), moveTotalDF[7:])
ax2.set_xticks(np.arange(0,9,1), moveTotalDF.index[7:].to_list(),rotation=45)

fig.suptitle('[대구광역시 ==> 타도 이동인구수]', fontsize=20)
plt.show()
```

```python
# - 데이터 추출
moveTotalDF=cityDF.sum()
print(moveTotalDF)
```

```python
def drawGraphe():
    fig=plt.figure(figsize=(12,10))
    ax1, ax2=fig.subplots(2,1, sharex=True)

    ax1.plot(moveTotalDF)
    ax2.bar(moveTotalDF.index.to_list(),moveTotalDF)

    fig.suptitle('[대구광역시 ==> 타도 이동인구수]', fontsize=20)
    plt.show()
```

```python
drawGraphe()
```
