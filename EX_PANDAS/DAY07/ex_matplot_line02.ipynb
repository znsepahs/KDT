### [남북한발전전력량 분석 및 시각화]
- 데이터셋 : 남북한발전전력량.xlsx
- 데이터구성
    - 1990년 ~ 2016년까지 수력, 화력, 원자력, 신재생 에너지 데이터
- 분석 : 남한과 북한의 에너지 데이터 변화 분석

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [2] 데이터 준비
FILE_DATA=r'C:\KDT\EX_PANDAS\DATAS\남북한발전전력량.xlsx'
```

```python
# [3] Excel => DataFrame 형태로 저장
dataDF=pd.read_excel(FILE_DATA)
```

```python
# [4] 데이터 확인
dataDF.info()
```

```python
# [5] 데이터 전처리 및 가공
# - 멀티인덱싱 : 컬럼 => 행인덱스 설정
# - (1) NaN컬럼에 대한 처리
# - (2) 두개의 컬럼을 행인덱스로 설정
```

```python
# - (1) NaN컬럼에 대한 처리
#dataDF[dataDF.columns[0]].fillna(method='ffill')
dataDF[dataDF.columns[0]].ffill(inplace=True)
```

```python
dataDF.head()
```

```python
# - (2) 두 개의 컬럼을 행 인덱스로 설정
dataDF.set_index(dataDF.columns[:2].to_list(),inplace=True)
print(f'[확인]\n{dataDF.head(2)}')
```

```python
# - 행 인덱스 names 속성 수정
idx_names=dataDF.index.names
print(f'idx_names :{idx_names}')

dataDF.index.set_names({idx_names[0] : '구분',idx_names[1]:'전력'},inplace=True)
print(f'[확인]\n{dataDF.head()}')
```

```python
# [5] 데이터 전처리 및 가공
# - '-' 데이터 값 <NA>로 치환 => 0으로 치환
dataDF.replace('-',pd.NA, inplace=True)
print(f'[확인]\n{dataDF.head()}')
```

```python
# - <NA> 데이터 값 0으로 치환
dataDF.isna().T.sum()
```

```python
dataDF.fillna(0,inplace=True)
dataDF.isna().T.sum()
```

```python
# - object => int16 형변환
dataDF=dataDF.astype('int16')
dataDF.dtypes
```

```python
# 한글폰트 설정 => 폰트 매니저 모음
from matplotlib import font_manager as fm
from matplotlib import rc

# 적용할 폰트 파일
FONT_FILE=r'C:\Windows\Fonts\NanumGothic.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=FONT_FILE).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# [6] 남한과 북한 발전량 합계 비교
plt.plot(dataDF.loc[('남한','합계')],'bo-',label='남한 - 합계')
plt.plot(dataDF.loc[('북한','합계')],'ro-',label='북한 - 합계')
plt.title("[남한 & 북한 발전량]")
plt.xlabel("YEAR")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
def drawLineGraph(title, columns, labels, fmfs, xlabel, ylabel):
    for idx, col in enumerate(columns):
        plt.plot(dataDF.loc[col],fmfs[idx])
    
    plt.title(f'---[{title}]---')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```

```python
drawLineGraph("Test",
              [('남한','수력'),('북한','수력'),('남한','합계'),('북한','합계')],
              ['남한-수력','북한-수력','남한-합계','북한-수력'],
              ['bo--','r^-','g+-','ko--'],
              '년도','발전량')
```

- 하나의 figure에 여러개 그래프 넣기

```python
# - figure 객체 생성
# - 생성된 figure 객체 안에 subplot 객체 생성
#   * 그래프 표시 방식 지정 : 가로/세로 몇개의 그래프를 넣을 지 지정
# - subplot에 그래프 넣기
```

```python
# => 하나의 Figure에서 수력/화력/원자력/합계 그래프 넣기
fig=plt.figure()

# - add_subplot(행수, 열수, 번호)
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

# - 각 subplot에 그래프 넣기
ax1.plot(dataDF.loc[('남한','수력')],'bo-',label='남한 - 수력')
ax2.plot(dataDF.loc[('남한','화력')],'ro-',label='남한 - 화력')
ax3.plot(dataDF.loc[('남한','원자력')],'yo-',label='남한 - 원자력')
ax4.plot(dataDF.loc[('남한','합계')],'ko-',label='남한 - 합계')

# - 각 subplot 별로 설정
ax1.set_title('남한 수력발전량')
ax1.set_xlabel('년도')

ax2.set_title('남한 화력발전량')
ax2.set_xlabel('년도')

ax3.set_title('남한 원자력발전량')
ax3.set_xlabel('년도')

ax4.set_title('남한 합계발전량')
ax4.set_xlabel('년도')

# - 화면에 띄우기
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# - subplot 객체 생성 : subplots() => axes 객체를 배열에 담아서 반환
fig=plt.figure(figsize=(14,8))

axes=fig.subplots(1,4)

# - subplot에 그래프 넣기
datas=dataDF.index.to_list()[:4]
fmfs=['bo-','go-','ro-','ko--']
for idx, ax in enumerate(axes):
    ax.plot(dataDF.loc[datas[idx]],fmfs[idx])

# - 전체 창 즉, figure 설정
fig.suptitle('남한의 에너지 발전량 변화 동향', fontsize=12)

# - 화면에 띄우기
plt.tight_layout()
plt.show()
```
