### [결측치 처리]
- 빈 칸 즉, 데이터가 없는 부분을 의미
- 결측치 체크 후 처리 방법 결정
- 처리 방법
    - 제거
    - 치환


```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# [2] 데이터 준비
file_name='data.csv'
```

```python
# [3] CSV => DataFrame 형태로 저장
dataDF=pd.read_csv(file_name)
```

```python
# [4] 데이터 기본 정보 확인
dataDF.info()
```

```python
# - 실제 데이터와 컬럼별 타입 체크
dataDF.head()
```

```python
# [5] 데이터 전처리
# [5-1] 결측치 체크 => isnull(),isna() 메서드
dataDF.isna()
```

```python
dataDF.isna().sum()
```

```python
# [5-2] 결측치 처리 => 방법 1) 삭제 dropna() 메서드
# 기본값 : 행 방향, 행 안에 1개의 컬럼이라도 Na면 삭제
dataDF.dropna()
```

```python
# 설정값 : 열 방향, 열 안에 1개의 원소라도 Na면 삭제
dataDF.dropna(axis=1)
```

```python
# 설정값 : 모든 원소가 NaN일때 삭제 how='all'
dataDF.dropna(axis=0, how='all')
```

```python
# 설정값 : 정상 데이터가 지정된 개수 미만이면 삭제 : thresh=정수
dataDF.dropna(axis=0, thresh=4)
```

```python
# 설정값 : 특정 컬럼만 NaN여부 검사 진행 후 결정 : subset
# - 나이를 분석하고 싶음 => 중요한 컬럼 age
dataDF.dropna(axis=0, subset=['age'])
```

```python
# - 나이와 이름을 분석하고 싶음 => 중요한 컬럼 age
dataDF.dropna(axis=0, subset=['age','name'])
```

```python
# [5-2] 결측치 처리 => 방법 2) 대체/치환 fillna() 메서드
# - 대체값 : 원하는 값 또는 컬럼의 평균값/최빈값 등을 사용 가능
#           이전의 값 또는 이후의 값으로 대체
dataDF.fillna(0)
```

```python
dataDF.fillna({'name':'unknown','age':0})
```

```python
# - 이전 컬럼의 값으로 채우기
dataDF.ffill()
```

```python
# - 아래 컬럼의 값으로 채우기
dataDF.bfill()
```

```python
# - 나이의 결측치는 나이 평균으로 채우기
dataDF.fillna({'age':dataDF['age'].mean()}, inplace=True)
```

```python
# - 이름의 결측치는 최빈값으로 채우기
dataDF.fillna({'name':dataDF['name'].mode()},inplace=True)
```
