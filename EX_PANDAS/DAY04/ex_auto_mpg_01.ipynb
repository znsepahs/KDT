### CSV 데이터 파일 다루기
- auto_mpg 데이터 파일 살펴보기

```python
# [1] 모듈로딩
import pandas as pd
```

```python
# [2] 데이터 준비
file_name='auto_mpg.csv'
```

```python
# [3] CSV 데이터 => DF 형태로 저장
mpgDF=pd.read_csv(file_name)
mpgDF
```

```python
# [4] 데이터 확인
print(f'데이터 형태 : {mpgDF.shape[0]}, 열 개수 :{mpgDF.shape[1]}')
print(f'데이터 컬럼 : {mpgDF.columns}')
print(f'데이터 인덱스 : {mpgDF.index}')
print(f'데이터 타입 : \n{mpgDF.dtypes}')
```

```python
# [5] 데이터 추출
# [5-1] 70년대 자동차 데이터만 추출
# p1=mpgDF['model year']%70<10
p1=mpgDF['model year']//10==7
mpgDF[p1]
```

```python
# [5-2] 실린더 개수가 4개이하인 자동차 데이터만 
p2=mpgDF['cylinders']<=4
mpgDF[p2]
```

```python
# [5-3] 일본차가 아닌 자동차 데이터 추출
p3=(mpgDF['origin']==1) | (mpgDF['origin']==2)
mpgDF[p3]
```

```python
# [5-4] 80년대 유럽 자동차 중에서 실린더 개수가 6개 이상인 데이터 추출
p4=(mpgDF['model year']//10==8) & (mpgDF['origin']==2) & (mpgDF['cylinders']>=6)
mpgDF[p4]
```

- [데이터의 기본정보 학인]

    - 기본정보 => 인덱스, 컬럼, 컬럼별 데이터 개수 및 타입 => info()
    - 실제 데이터 일부 => head() / tail()
    - 컬럼별로 데이터 분포 => 기술통계 수치 describe()

```python
# [3] 컬럼별 데이터 분포 확인
mpgDF.describe(include='all')
```
