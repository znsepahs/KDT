### [ 데이터 표준화 ]
- 수집한 데이터를 동일한 형태로 맞추어 주는 것
- 단위변환, 자료형 변환

```python
# [1] 모듈로딩
import pandas as pd
import numpy as np
```

```python
# [2] 데이터 준비
file_name='auto_mpg.csv'
```

```python
# [3] CSV 파일 => DataFrame 형태로 저장
mpgDF=pd.read_csv(file_name)
```

```python
# [4] 데이터 학인
# - 기본 정보 확인
mpgDF.info()
```

```python
# - 컬럼과 실제 데이터 타입 확인
print(mpgDF.head(),mpgDF.tail(),sep='\n\n')
```

```python
# - 컬럼별 데이터 분포 확인
mpgDF.describe()
```

```python
# [5] 데이터 전처리
# [5-1] 결측치 체크
mpgDF.isna().sum()
```

```python
# [5-2] 중복데이터 체크
mpgDF.duplicated().sum()
```

```python
# [5-3] 단위 변환 표준화 mpg => kpl 컬럼 추가
mpgDF['kpl']=mpgDF.mpg*0.425144
mpgDF
```

```python
# [5-4] 자료형 변환 표준화
# horsepower object => int
mpgDF['horsepower'].value_counts()['?']
```

```python
# 데이터를 변경해주는 메서드 => replace
mpgDF['horsepower'].replace('?',np.nan,inplace=True)
mpgDF.isna().sum()
```

```python
mpgDF2=mpgDF.dropna().copy()
mpgDF2.info()
```

```python
mpgDF2['horsepower']=mpgDF2['horsepower'].astype('int64')
```

```python
mpgDF2.info()
```

```python
# [5-5] 컬럼별 데이터의 특성 확인 후 자료형 변환 표준화
# - 'cylinders' => 데이터를 분류/그룹 가능. 범주형
# - 'origin' => 데이터를 분류/그룹 가능. 범주형
mpgDF['cylinders']=mpgDF['cylinders'].astype('category')
mpgDF['origin']=mpgDF['origin'].astype('category')
mpgDF.info()
```

- 초기에는 시리즈 내에 있는 데이터를 가지고 알파벳 순서로 정렬이 되어 있다.
- custom order를 사용하고자 할 때는 cat.reorder_categories로 원하는 순서를 지정해주면 된다.

- 카테고리를 추가하고자 할 때에는 .cat.add_categories
- 삭제할 때에는 .cat.remove_categories
- 사용되지 않는 카테고리만 삭제할 때에는 .cat.remove_unused_categories
- 또한 .cat.as_ordered()를 통해 카테고리 데이터 간 대소비교도 확인할 수 있다.

```python
# [5-6] 컬럼별 데이터의 특성 확인 후 데이터 변환
# - 'horsepower' => 저출력-일반출력-고출력 => 범주형 : cut() 메서드
hp=pd.cut(mpgDF2['horsepower'],
          bins=3, 
          labels=['저출력','일반출력','고출력'])
```

```python
mpgDF2['horsepower_range']=hp
mpgDF2
```

- [문자열 데이터를 분리 시켜주는 기능 => 컬럼 분리]
    - 조건, 데이터가 str이어야 함!
    - 형식 : Series.str.메서드()

```python
namesSR=mpgDF2['car name'].str.split()
namesSR.head(3)
```

```python
# 나누어진 문자열을 하나씩 컬럼으로 분리
namesDF=mpgDF2['car name'].str.split(expand=True,n=2)
namesDF.info()
```

- [DataFrame 연결하기]
    * 함수 : pandas.concat([df1,df2...])

```python
DF3=pd.concat([mpgDF2,namesDF],axis=1)
DF3.info()
```

```python
# 행 방향 연결 시 인덱스 다시 정리 필요
# - 매개변수 ignore_index=True
DF4=pd.concat([mpgDF2,namesDF])
DF4.info()
```
