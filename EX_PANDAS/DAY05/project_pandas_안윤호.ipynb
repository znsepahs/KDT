### 전통주 종류 설명
- 탁주 : 전분질 원료와 국을 주원료로 하여 발효시킨 술덧(주요)을 혼탁하게 제성한 것을 말함.

- 약주 : 전분질 원료와 국을 주원료로 하여 발효시킨 술덧(주요)을 여과하여 제성한 것을 말함.

- 청주 : 전분질 원료와 국을 주원료로 하여 발효시킨 술덧(주요)을 여과 제성한 것 또는 발효 제성 과정에 주류 등을 첨가한 것을 말함.

- 과실주 : 과실 또는 과즙을 주 원료로 하여 발효시킨 술덧(주요)을 여과 제성한 것 또는 발효 과정에 과실, 당질 또는 주류 등을 첨가한 것을 말함.

- 일반증류주 : 전분질 또는 당분질을 주 원료로 하여 발효, 증류한 것, 증류주를 혼합한 것, 과실(포도 등 발효시킬 수 있는 과실 제외)의 추출액을 첨가한 것. 주정, 소주, 위스키, 브랜디 이외의 주류로서 주세법에서 규정한 것을 말함.

- 리큐르 : 전분질 또는 당분질을 주원료로 하여 발효시켜 증류한 주류에 인삼, 과실(포도 등 발효 시킬 수 있는 과실 제외) 등을 침출시킨 것이거나 발효 증류 제성과정에 인삼, 또는 주정, 소주, 일반증류주의 발효, 증류, 제성과정에 주세법에서 정한 물료를 첨가한 것을 말함.

- 기타주류 : 발포주가 포함되며, 따로 기준 및 규격이 제정되지 아니한 주류로서 주세법에서 규정한 것을 말함.

```python
# 사용할 데이터는 3개 / 기간은 2018~2022년 사이 / 19년대비 20년의 증감률을 따로 확인
# 1. 국내시장에서의 전통주 거래량에 대한 데이터
# 2. 전통주 시장 규모의 확대를 확인 할 수 있는 전통주 제조 면허 수 데이터
# 3. 전통주 수출 현황에 대한 데이터
```

```python
data1='전통주산업현황.xlsx' # 출처 : 국세청 통계연보
df1=pd.read_excel(data1)
df1 #단위 : Kl(킬로리터) , 백만원
```

```python
# 가시성을 위해 출고량과 출고금액을 나타내는 2개의 데이터 프레임으로 분리
KL=df1[['구분','2018출고량','2019출고량','2020출고량','2021출고량','2022출고량','19년대비증감률-출고량']]
KL # 2018~2022까지의 전통주 출고량
# 발표 주제에 어울리는 데이터는 출고량으로 판단함
```

```python
Price=df1[['구분','2018출고금액','2019출고금액','2020출고금액','2021출고금액','2022출고금액','19년대비증감률-출고금액']]
Price # 2018~2022까지의 전통주 출고금액
```

```python
data2='전통주제조면허현황.xlsx' # 출처 : 국세청 통계연보
df2=pd.read_excel(data2)
df2 #단위 : 개
```

```python
data3='전통주수출현황.xlsx' # 출처 : 관세청 수출입현황 통계
df3=pd.read_excel(data3)
df3 #중량 : 톤 / 금액 : 1,000$
```

```python
# 2020년의 증감률이 필요해서 증감률을 계산하고 DataFrame에 새로운 열로 추가
a=df3['2020중량']
b=df3['2019중량']
ton20=((a-b)/b)*100
Ton20=ton20.to_frame()
df3['19년대비증감률-중량(수출)']=Ton20
df3 # 19년대비 수출 전통주 중량 증감률
```

```python
sales20=((df3['2020금액']-df3['2019금액'])/df3['2019금액'])*100
Sales20=sales20.to_frame()
df3['19년대비증감률-금액(수출)']=Sales20
df3 # 19년대비 수출 전통주 금액 증감률
```

```python
# 가시성을 위해 각각 중량과 금액 정보를 갖는 2개의 데이터 프레임으로 분리
Drink_ton=df3[['구분','2018중량','2019중량','2020중량','2021중량','2022중량']]
Drink_ton # 중량 / 단위 : 톤
```

```python
Drink_price=df3[['구분','2018금액','2019금액','2020금액','2021금액','2022금액']]
Drink_price # 수출액 / 단위 : 1,000$
```
